```python
df=df.drop('Year',axis=1)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

df['Director(s)']=label_encoder.fit_transform(df['Director(s)'])
df['Film']=label_encoder.fit_transform(df['Film'])

```

```python
X=df.drop('Nomination/Winner',axis=1)
y=df['Nomination/Winner']
```

```python
from sklearn.model_selection import train_test_split  
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.metrics import f1_score


model=KNeighborsClassifier(n_neighbors=11,metric='euclidean')
pipeline=make_pipeline(model)
from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(estimator = pipeline,
                             X = X,
                             y = y,
                             scoring = 'accuracy',
                             cv = 20)

accuracies
```

```python
accuracies.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='entropy',random_state=0)
clf.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
y_pred = clf.predict(X_test)
y_train = clf.predict(X_train)
from sklearn.metrics import accuracy_score 
print("Train Accuracy is:",accuracy_score(y_train,y_train))
print("Test Accuracy is:",accuracy_score(y_test,y_pred))
```

```python
cr =classification_report(y_test,y_pred)
print("Classification Report")
print(cr)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score 
svc= SVC(kernel='linear',C=0.1)
svc.fit(X_train,y_train)

from sklearn.metrics import accuracy_score 
print("Train Accuracy is:",accuracy_score(y_train,y_train))
print("Test Accuracy is:",accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
svc=SVC(kernel='linear',C=0.1)
scores = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
print(scores)
```
